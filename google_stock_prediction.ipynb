```python
import pandas as pd
import numpy as np
```

```python
a=pd.read_csv('/content/indexData.csv')
a
```

```python
a.columns
```

```python
a.shape
```

```python
a.info()
```

```python
a.isnull().sum()
```

```python
a.fillna(method='ffill',inplace=True)
```

```python
a.isnull().sum()
```

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
a['Date']=label_encoder.fit_transform(a['Date'])
a['Date'].unique()
```

```python
x=a[['Date', 'Open', 'High', 'Low', 'Close','Adj Close']]
x
```

```python
y=a[['Volume']]
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytrain.shape
```

```python
ytest.shape
```

```python
from sklearn.metrics import r2_score
from sklearn import linear_model
b = linear_model.Lasso(alpha=0.1)
b.fit(xtrain,ytrain)
ypred=b.predict(xtest)
score=r2_score(ytest,ypred)
print(score)
```

```python
prediction = b.predict([[261,531.119995,531.119995,531.119995,531.119995,531.119995]])
print(prediction)
```
